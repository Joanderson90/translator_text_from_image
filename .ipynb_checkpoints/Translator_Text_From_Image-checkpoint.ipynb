# Show the Image

```python

image_path = 'Example_Text_Portuguese.JPG'

#Open the image using PIL.
image = PIL.Image.open(image_path)

image
```

# Read the Image Using EasyOCR


```python

#Here, we put language codes in list format and pass it as  argument. In this case, it is ['en', 'pt', 'es'].
#See all languages available on EasyOCR -> https://www.jaided.ai/easyocr/

reader = easyocr.Reader(['en', 'pt', 'es'])

#Here, we extract the text read from the image.
#Obs:The time to extract the text depends on the size of the image. Big images generally take more time.

text_read = reader.readtext(image_path, detail = 0, paragraph = True)

text_read
```

# Concatenate the extracted text from Image

```python

text_concatenated = ''

for text in text_read:
    
    text_concatenated += text

text_concatenated
```

# Translate the extracted text

```python

language_my_text = 'pt'  #Text language from Image imported.
language_to_translate = 'en'  #The language that you want to translate.

#Here, we translate the text using the translators. 

text_translated = ts.google(text_concatenated, from_language = language_my_text, to_language= language_to_translate )
text_translated
```
